```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
next_button = True
page_num = 1
movie_names= []
year = []
rating = []
critics_consensus = []
synopsis = []
director = []
images = []
cast_lst = []
while next_button:
    base_site = "https://editorial.rottentomatoes.com/guide/essential-movies-to-watch-now/"+str(page_num)
    response = requests.get(base_site)
    html = response.content
    soup = BeautifulSoup(html,"html.parser")
    next_button = soup.find(class_="post-page-numbers")

    for i in soup.find_all(class_="article_movie_title"):
        movie_names.append(i.find("h2").find("a").text)
        year.append(i.find(class_="subtle").text.strip("()"))
        rating.append(i.find(class_="tMeterScore").text)
    for i in soup.find_all(class_="info critics-consensus"):
        critics_consensus.append(i.text.split(": ")[1])
    for i in soup.find_all(class_="info synopsis"):
        synopsis.append(i.text.split(": ")[1].strip("... [More]"))
    for i in soup.find_all(class_="info director"):
        director.append(i.find("a").text)
    for i in soup.find_all(class_="article_poster"):
        images.append(i.get("src"))
    for i in soup.find_all(class_="info cast"):
        a=i.find_all("a")
        temp_lst = []
        for j in a :
            temp_lst.append(j.text)
        cast_lst.append(temp_lst)
        
    page_num+=1
```

```python
df = pd.DataFrame({"Movie":movie_names,
                  "Rating":rating,
                  "Year":year,
                  "Critics_Consensus":critics_consensus,
                  "Synopsis":synopsis,
                  "Starring":cast_lst,
                  "Director":director,
                  "Image":images})
df
```

```python
df.to_excel("Rotten Tomatoes 300 Movies to watch.xlsx",index = False)
```
